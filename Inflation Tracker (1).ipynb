```python
# Calculate Year-over-Year (YoY) percentage changes
yoy_changes = data.iloc[:, 3:].pct_change(axis=1) * 100

# Assign these changes back to the dataframe
data_yoy = data.copy()
data_yoy.iloc[:, 3:] = yoy_changes

# Display the head of the new dataset with YoY changes
data_yoy.head()
```

```python
# Calculate 5-year moving averages for each row, ignoring non-numeric columns
moving_avg_5yr = data.iloc[:, 3:].rolling(window=5, axis=1).mean()

# Assign these averages back to the dataframe
data_ma = data.copy()
data_ma.iloc[:, 3:] = moving_avg_5yr

# Display the head of the new dataset with 5-year moving averages
data_ma.head()
```

```python
# Combine original data with YoY changes and moving averages
data_combined = data.copy()

# Add '_yoy' and '_ma' suffixes to differentiate columns
data_combined.iloc[:, 3:] = data.iloc[:, 3:]
data_combined_yoy = yoy_changes.add_suffix('_yoy')
data_combined_ma = moving_avg_5yr.add_suffix('_ma')

# Concatenate original, YoY, and moving average data
data_combined = pd.concat([data, data_combined_yoy, data_combined_ma], axis=1)

# Display the combined dataset head
data_combined.head()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Calculate correlation matrix
corr_matrix = data_combined.corr()

# Plot the heatmap
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
import numpy as np

# Plot histograms for YoY changes and 5-year moving averages
data_combined_yoy.hist(figsize=(15, 10), bins=20)
plt.show()

data_combined_ma.hist(figsize=(15, 10), bins=20)
plt.show()
```

```python
countries_to_plot = ['USA', 'IND', 'CHN']  # example countries

for country in countries_to_plot:
    country_data = data_combined[data_combined['Country Code'] == country].iloc[:, 3:]
    country_data.T.plot(figsize=(10, 6), title=f"Inflation Trends - {country}")
    plt.show()
```

```python
# Select the relevant columns for a single country
ind_data = data[data['Country'] == 'India'].iloc[:, 3:].T

# Rename the column for clarity
ind_data.columns = ['Inflation Rate']

# Reset the index to ensure the years are the index
ind_data.index = ind_data.index.astype(int)
ind_data = ind_data.sort_index()

# Preview the inflation time series
ind_data.head()
```

```python
from statsmodels.tsa.stattools import adfuller

# Perform ADF test
adf_test = adfuller(ind_data['Inflation Rate'])

print(f"ADF Statistic: {adf_test[0]}")
print(f"p-value: {adf_test[1]}")
```

```python
#p value< 0.05, hence the series is stationary
```

```python
from pmdarima import auto_arima

# Automatically select the best p, d, q for ARIMA
auto_arima_model = auto_arima(ind_data, seasonal=False, stepwise=True, trace=True)

# Check the best parameters
auto_arima_model.summary()
```

```python
from statsmodels.tsa.arima.model import ARIMA

# Fit the ARIMA model
model = ARIMA(ind_data, order=(2,1,1))  # Replace p, d, q with values from auto_arima
model_fit = model.fit()

# Forecast for the next 5 years
forecast = model_fit.forecast(steps=5)
print(forecast)
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the range for plotting (focus on the last few years and future forecast)
n_years = 50  # Number of years to show for historical data
end_year = len(ind_data) + len(forecast)  # Total length with forecast included

# Create the figure
plt.figure(figsize=(12, 6))

# Plot the historical inflation data (last N years)
plt.plot(ind_data[-n_years:], label='Historical Inflation', color='blue', linewidth=2)

# Plot the forecasted inflation data with a different color
plt.plot(range(len(ind_data), end_year), forecast, label='Forecasted Inflation', color='red', linestyle='--', linewidth=2)

# Highlight the forecast period with a shaded area
plt.axvspan(len(ind_data), end_year, color='gray', alpha=0.1, label='Forecast Period')

# Add confidence intervals (if available)
# Lower and upper confidence intervals from ARIMA model (replace with actual if available)
# conf_int = arima_fit.conf_int() 
# plt.fill_between(range(len(us_data), end_year), conf_int[:, 0], conf_int[:, 1], color='red', alpha=0.2)

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Inflation Rate')
plt.title('ARIMA Inflation Forecast with Historical Data')

# Add a vertical line to separate historical data from the forecast
plt.axvline(x=len(ind_data), color='black', linestyle=':', label='Forecast Start')

# Add a legend and grid for clarity
plt.legend()
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the number of years to forecast
forecast_horizon = len(forecast)  # Number of years in the forecast (e.g., 5)

# Define the last available year in the historical data (e.g., 2022)
last_year = 2022

# Create an array of years for the historical data (based on the years in your dataset)
historical_years = np.arange(last_year - len(ind_data) + 1, last_year + 1)

# Create an array of future years for the forecasted data (2023, 2024, ..., etc.)
forecast_years = np.arange(last_year + 1, last_year + 1 + forecast_horizon)

# Create the figure
plt.figure(figsize=(12, 6))

# Plot the historical inflation data (last N years)
plt.plot(historical_years, ind_data[-len(historical_years):], label='Historical Inflation', color='blue', linewidth=2)

# Plot the forecasted inflation data
plt.plot(forecast_years, forecast, label='Forecasted Inflation', color='red', linestyle='--', linewidth=2)

# Highlight the forecast period with a shaded area
plt.axvspan(last_year, last_year + forecast_horizon, color='gray', alpha=0.1, label='Forecast Period')

# Add labels and title
plt.xlabel('Year')
plt.ylabel('Inflation Rate')
plt.title('ARIMA Inflation Forecast with Historical Data')

# Add a vertical line to separate historical data from the forecast
plt.axvline(x=last_year, color='black', linestyle=':', label='Forecast Start')

# Add a legend and grid for clarity
plt.legend()
plt.grid(True)

# Show the plot
plt.tight_layout()
plt.show()
```
