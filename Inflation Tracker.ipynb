```python
data=data.drop(columns=['Note', 'Unnamed: 59', 'Unnamed: 60', 'Unnamed: 61', 'Unnamed: 62', 'Unnamed: 63'], errors='ignore')
```

```python
data = data[data['Series Name'] == 'Headline Consumer Price Inflation']
```

```python
data = data.drop(columns=['IMF Country Code','Series Name'])
data.head()
```

```python
missing_values = data.isnull().sum()
print(missing_values) 
```

```python
# Replace NaN values with the mean inflation rate for each country
data_filled = data.copy()

# Calculate the mean for each country across the inflation columns (ignoring NaN)
data_filled.iloc[:, 3:] = data.iloc[:, 3:].apply(lambda row: row.fillna(row.mean()), axis=1)

# Check the result
data_filled.head()
```

```python
data = data_filled.copy()
```

```python
# List of top 10 economies by country name
top_10_economies = ['United States', 'China', 'Japan', 'Germany', 'India', 
                    'United Kingdom', 'France', 'Brazil', 'Italy', 'Canada']

# Filter the dataset based on these countries
top_10_economies_data = data[data['Country'].isin(top_10_economies)]

# Display the head of the new dataset
top_10_economies_data.head(10)
```

```python
import matplotlib.pyplot as plt
```

```python
#Distribution of Inflation Rates
inflation_columns = data.columns[3:]  # Assuming first three columns are Country Code, Country, Indicator Type
plt.figure(figsize=(12, 8))
data[inflation_columns].T.boxplot()
plt.xticks(rotation=90)
plt.title('Boxplot of Inflation Rates Over the Years')
plt.ylabel('Inflation Rate')
plt.xlabel('Year')
plt.show()
```

```python
import seaborn as sns
```

```python
#Correlation Analysis
correlation = data[inflation_columns].corr()
plt.figure(figsize=(12, 10))
sns.heatmap(correlation, annot=True, fmt='.2f', cmap='coolwarm', square=True)
plt.title('Correlation Heatmap of Inflation Rates')
plt.show()
```

```python
#Yearly Analysis for India
country_data = data[data['Country'] == 'India'].iloc[:, 3:].T
country_data.columns = ['Inflation Rate']
plt.figure(figsize=(12, 6))
plt.plot(country_data.index, country_data['Inflation Rate'], marker='o')
plt.title('Inflation Rate Over the Years for India')
plt.xlabel('Year')
plt.ylabel('Inflation Rate')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

```python
# Calculate Year-over-Year (YoY) percentage changes
yoy_changes = data.iloc[:, 3:].pct_change(axis=1) * 100

# Assign these changes back to the dataframe
data_yoy = data.copy()
data_yoy.iloc[:, 3:] = yoy_changes

# Display the head of the new dataset with YoY changes
data_yoy.head()
```

```python
# Calculate 5-year moving averages for each row, ignoring non-numeric columns
moving_avg_5yr = data.iloc[:, 3:].rolling(window=5, axis=1).mean()

# Assign these averages back to the dataframe
data_ma = data.copy()
data_ma.iloc[:, 3:] = moving_avg_5yr

# Display the head of the new dataset with 5-year moving averages
data_ma.head()
```

```python
# Combine original data with YoY changes and moving averages
data_combined = data.copy()

# Add '_yoy' and '_ma' suffixes to differentiate columns
data_combined.iloc[:, 3:] = data.iloc[:, 3:]
data_combined_yoy = yoy_changes.add_suffix('_yoy')
data_combined_ma = moving_avg_5yr.add_suffix('_ma')

# Concatenate original, YoY, and moving average data
data_combined = pd.concat([data, data_combined_yoy, data_combined_ma], axis=1)

# Display the combined dataset head
data_combined.head()
```
